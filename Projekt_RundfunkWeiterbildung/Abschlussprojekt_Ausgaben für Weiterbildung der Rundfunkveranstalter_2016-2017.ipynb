Projekt "Entwicklung der Ausgaben für Weiterbildung bei den Radio- und Fernsehveranstaltern": Daten für die Jahre 2016 - 2017

Ziel des Projekts ist es zu zeigen, welcher (Typ von) Rundfunkveranstalter am meisten in die Weiterbildung seiner Angestellten investiert. In einem zweiten Schritt können die Ausgaben in Bezug zu anderen Kennzahlen (u.a. Reichweite) gesetzt werden.

Datenquellen: (1) Jahresberichte der Veranstalter (seit 2008 auf BAKOM-Website, für 2015 und 2016 als Excel-Sheet vom BAKOM erhalten. (2) Reichweitenmessungen von Mediapulse AG.
