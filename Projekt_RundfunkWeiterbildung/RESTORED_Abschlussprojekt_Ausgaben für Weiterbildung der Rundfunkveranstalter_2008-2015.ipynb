Und jetzt alles zusammenbauen, so dass über alles Veranstalter iteriert wird.
- über die links je den table sammeln
- dann den Namen in dic schreiben
- dann alle zeilen bzw. links rausziehen, die "Jahresbericht" im Namen haben.
==> Ziel: Für all diejenigen Veranstalter, zu denen Jahresberichte vorhanden sind, eine Linkliste zu den Jahresberichten zusammenstellen.

```python
# Den Tabellen Text und den Html Code pro Veranstalter in ein dict speichern
driver = webdriver.Chrome('/usr/local/bin/chromedriver')

veranstalter_liste = []

for link in links:

    try:
        driver.get(link)
        time.sleep(12)
        table = driver.find_element_by_tag_name("table")
        table_text = table.text
        source = driver.page_source
        veranstalter_liste.append({"table_text": table_text, "html_source": source})
        print("success: ", link)
        print("Table: ", table_text[:15]) 
        print("Source: ", source[:15])
        print()
    except Exception as exception:
        print("Failed to get link: " + link + "\n" + exception)
        continue
    
len(veranstalter_liste)
```

```python
# Von der Veranstalter liste mit den dicts (mit tabellen text und html code) die Daten auslesen
# und in ein neues dict veranstalerinfo_list als einzelne dicts speichern
veranstalterinfo_list = []

for eintrag in veranstalter_liste:
    if len(eintrag["table_text"].split("\n")) <= 1:
        continue

    zeilen = eintrag['table_text'].split("\n")
    daten = {"programmname": zeilen[0]}

    
    # Hier habe ich ein paar probleme gefunden und deswegen gibts hier neuen code. 
    # Medientyp ist nicht immer in zeile 2 deswegen machen wir jetzt hier ein loop und
    # ordnen hier die sachen "manuel" zu.
    
    # Wir brauche einen Shift um zu sehen, ab wo wir weiterarbeiten können
    offset = 0
    for idx, zeile in enumerate(zeilen[1:]):
        zeilen_anfang = zeile.split()

        if len(zeilen_anfang) > 1:
            if zeilen_anfang[0] == "Medientyp":
                daten["medien_typ"] = zeile
            elif zeilen_anfang[0] == "Kategorie":
                daten["kategorie"] = zeile
                offset = idx + 1
        
    # Die Adresse müsste jetzt in der 2. Zeile nach Kategorie sein.
    # Also Shift (sollte die Zeile von Kategorie sein) + 2
        
    shift = 0 # muss hier definiert werden
    daten["adresse"] = zeilen[shift + 2]
    daten["ort"] = zeilen[shift + 3]
    
    # Und jetzt suchen wir noch schnell die Verbreitung und die URL
    # und telefonnummer
    # Aus faulheit einfach im ganzen text ohne slicing.
    
    for zeile in zeilen:
        if zeile[:3] == "+41":
            daten["tel"] = zeile
        if zeile[:4] == "http":
            daten["url"] = zeile
        if len(zeile.split()) > 1 and zeile.split()[0] == "Verbreitung":
            daten["verbreitung"] = zeile
            
    # Jetzt noch die Links
    daten["jahresberichte"] = []
    soup = BeautifulSoup(eintrag["html_source"], 'html.parser')
    links = soup.find_all("a", {"title": re.compile("Jahresbericht")})

    for link in links:
        daten["jahresberichte"].append('https://rtvdb.ofcomnet.ch' + link.attrs["href"])

    
    # Und in die liste speichern
    veranstalterinfo_list.append(daten)

# Als Ergebnis sollte nachfolgen eigentlich nur die Fehler aufgelistet werden
"""
for table in tables:
    for idx, line in enumerate(table.split("\n")):
        liste = print(idx, line)
        veranstalterinfo_list.append(liste)
"""
```

```python
# Schauen was so rausgekommen ist.

for eintrag in veranstalterinfo_list:
    for key, value in eintrag.items():
        print(str(key) + ": " + str(value))
    print("\n")
```
