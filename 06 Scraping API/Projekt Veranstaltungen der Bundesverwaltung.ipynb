```python
#for-loop bauen, damit wir eine Tabelle erhalten mit den Spalten Ort, Datum, Text
```

```python
veranstaltungen = []

ereignisse = list(range(6000,6005))#hier definieren, welche Fälle ausgelesen werden sollen.

for wert in ereignisse:
    url_ohne_ereignisnummer = "https://www.admin.ch/gov/de/start/dokumentation/veranstaltungen.event-id-"
    urlende = ".html"
    url = url_ohne_ereignisnummer+str(wert)+urlende
    #print(url)
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    
    try: #hinweis: wenn es klapp, wenn es entsprechende Daten zum Abgreifen gibt, dann mach weiter
        ort = soup.find_all('td', {'class': ""})[0].text.replace('\t','').replace('\xa0',' ').replace('\n',' ').strip()                             
        datum = soup.find_all('td', {'class': ""})[1].text.replace('\n', ' ').replace('\t','').strip()
        mitteilung = soup.find_all('td', {'class': ""})[2].text.replace('\n', ' ').replace('\t','').strip()

        minidict = {'Ort': ort,
                    'Datum': datum,
                    'Mitteilung': mitteilung}
        veranstaltungen.append(minidict)
    except:
        (print(wert))#Fehler werden ausgedruckt
```

```python
pd.DataFrame(veranstaltungen)
```
